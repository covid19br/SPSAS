# SIR
$S+I+R=N$

$s=S/N, i=I/N, r=R/N$

$s+i+r=1$

$ \frac{ds}{dt} = f(s,i,r) = -\beta \cdot s \cdot i $

$ \frac{di}{dt} = g(s,i,r) = \beta \cdot s \cdot i - \gamma \cdot i$

$ \frac{dr}{dt} = h(s,i,r) = \gamma \cdot i $

```python
%matplotlib inline

import numpy as np
from matplotlib import pyplot as pt
from scipy.integrate import odeint

x0=np.array([0.9999,0.0001,0]) #s,i,r
t=np.arange(0,30,0.1)
beta = 4
gamma = 1

def SIR(x,t,beta,gamma):
    return np.array([ -beta*x[0]*x[1],              #ds/dt
                       beta*x[0]*x[1]-gamma*x[1],   #di/dt
                       gamma*x[1] ])                #dr/dt

x =  odeint(SIR, x0, t, (beta, gamma)) #R0=4 com beta=4, gamma=1
x2 = odeint(SIR, x0, t, (2, 1))        #R0=2 com beta=2, gamma=1
x3 = odeint(SIR, x0, t, (4, 2))        #R0=2 com beta=4, gamma=2

pt.figure(1)
pt.plot(t, x)
pt.xlabel('t')
pt.ylabel('x=[s,i,r]')
pt.legend(['s', 'i', 'r'], loc='best')
print(x[-1])

pt.figure(2)
pt.plot(t, x2)
pt.xlabel('t')
pt.ylabel('x2=[s,i,r]')
pt.legend(['s', 'i', 'r'], loc='best')
print(x2[-1])

pt.figure(3)
pt.plot(t, x3)
pt.xlabel('t')
pt.ylabel('x3=[s,i,r]')
pt.legend(['s', 'i', 'r'], loc='best')
print(x3[-1])
```

#Retrato de Fase

$ i(s)=1-s+\frac{1}{R_0}ln(\frac{s}{s_0}) $

$ i(s)=1-s+0,25ln(\frac{s}{0,999}) $ #R0=4, s0=0,999


```python
from matplotlib import pyplot as pt
from scipy.integrate import odeint
import numpy as np

x0=np.array([0.999,0.001,0]) #s0,i0,r0
t=np.arange(0,15,0.1)
beta = 2
gamma = 1

def SIR(x,t,beta,gamma):
    return np.array([ -beta*x[0]*x[1],              #ds/dt=f(s,i,r)
                       beta*x[0]*x[1]-gamma*x[1],   #di/dt=g(s,i,r)
                       gamma*x[1] ])                #dr/dt=h(s,i,r)

x =  odeint(SIR, x0, t, (beta, gamma))              #(s,i,r) com (s(0),i(0),r(0))=x0

s=np.arange(0.2,1,0.01)

pt.figure(1) #quando usei 0.1 ou mesmo beta=4 a figura ficou estranha
pt.plot(x[0,0], x[0,1], 'o')                        #plota o ponto inicial
pt.plot(x[:,0], x[:,1])                             # i em função de s
pt.plot(s,1-s+(gamma/beta)*np.log(s/0.999),ls=':')
pt.xlabel('s')
pt.ylabel('i')
pt.legend(['Ponto inicial', '$i(s)$ numericamente','$i(s)$ com a fórmula'], loc='best')
```

```python
from matplotlib import pyplot as pt
from scipy.integrate import odeint

x0=np.array([0.9999,0.0001,0]) #s,i,r
t=np.arange(0,15,0.1)
beta = 4
gamma = 1

def SIR(x,t,beta,gamma):
    return np.array([ -beta*x[0]*x[1],              #ds/dt=f(s,i,r)
                       beta*x[0]*x[1]-gamma*x[1],   #di/dt=g(s,i,r)
                       gamma*x[1] ])                #dr/dt=h(s,i,r)

x =  odeint(SIR, x0, t, (beta, gamma))              #(s,i,r) com (s(0),i(0),r(0))=x0
x2 = odeint(SIR, [0.95, 0.05,0], t, (beta, gamma))    # mudando a condição inicial
x3 = odeint(SIR, [0.9, 0.0001,0], t, (beta, gamma))


pt.figure(1)
pt.plot(x[0,0], x[0,1], 'o')                        #plota o ponto inicial
pt.plot(x[:,0], x[:,1])                             # i em função de s
pt.xlabel('s')
pt.ylabel('i')
pt.plot(x2[0,0], x2[0,1], 'og')                     #plota o novo ponto inicial
pt.plot(x2[:,0], x2[:,1])                           # i em função de s (para o novo x0)
pt.plot(x3[0,0], x3[0,1], 'or')
pt.plot(x3[:,0], x3[:,1])

pt.figure(2)
S, I = np.meshgrid(np.arange(0.01, 1.02, .1), np.arange(0.01, 0.42, 0.1)) # cria uma matriz de pontos
dy = SIR(np.array([S, I]), 0, beta, gamma) # calcula ds/dt e di/dt nos pontos das matrizes
pt.quiver(S, I, dy[0,:], dy[1,:], scale_units='xy', angles='xy')
pt.plot(x[0,0], x[0,1], 'o')
pt.plot(x[:,0], x[:,1])
pt.xlabel('s')
pt.ylabel('i')

```

```python
from matplotlib import pyplot as pt
from scipy.integrate import odeint

x0=np.array([0.999,0.001,0]) #s,i,r
t=np.arange(0,15,0.1)
beta = 4
gamma = 1

def SIR(x,t,beta,gamma):
    return np.array([ -beta*x[0]*x[1],              #ds/dt=f(s,i,r)
                       beta*x[0]*x[1]-gamma*x[1],   #di/dt=g(s,i,r)
                       gamma*x[1] ])                #dr/dt=h(s,i,r)

x =  odeint(SIR, x0, t, (beta, gamma))              #(s,i,r) com (s(0),i(0),r(0))=x0
x2 = odeint(SIR, [0.95,0.001,0.049], t, (beta, gamma))    # mudando a condição inicial
x3 = odeint(SIR, [0.85, 0.001,0.149], t, (beta, gamma))

#plota o ponto inicial
pt.plot(x[0,0], x[0,1], 'o')
pt.plot(x2[0,0], x2[0,1], 'og')
pt.plot(x3[0,0], x3[0,1], 'or')

pt.axvline(gamma/beta,c='k',ls=':',label='gamma/beta')

pt.plot(x[:,0], x[:,1])                     # i em função de s
pt.plot(x2[:,0], x2[:,1])                   # i em função de s (para o novo x0)
pt.plot(x3[:,0], x3[:,1])

pt.xlabel('s')
pt.ylabel('i')
pt.legend(['s0=0.999', 's0=0.95', 's0=0.85','s=1/R0'], loc='best')

pt.axvline(0,c='k',label='i')
pt.axhline(0.,c='k',label='s')
```

```python
from matplotlib import pyplot as pt
from scipy.integrate import odeint

x0=np.array([0.999,0.001,0]) #s,i,r
t=np.arange(0,15,0.1)
beta = 4
gamma = 1

def SIR(x,t,beta,gamma):
    return np.array([ -beta*x[0]*x[1],              #ds/dt=f(s,i,r)
                       beta*x[0]*x[1]-gamma*x[1],   #di/dt=g(s,i,r)
                       gamma*x[1] ])                #dr/dt=h(s,i,r)

x =  odeint(SIR, x0, t, (beta, gamma))              #(s,i,r) com (s(0),i(0),r(0))=x0

#plota o ponto inicial
pt.plot(x[-1,0], x[-1,1], 'o')

pt.axvline(gamma/beta,c='k',ls=':',ymin=0.05,label='gamma/beta')

pt.plot(x[:,0], x[:,1])                     # i em função de s

pt.xlabel('s')
pt.ylabel('i')
pt.legend(['$s_{\infty}$', '$s=1/R_0$'], loc='best')

pt.axvline(0,c='k',label='i')
pt.axhline(0.,c='k',label='s')
```
